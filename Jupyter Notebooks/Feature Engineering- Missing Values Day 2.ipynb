#### Random Sample Imputation

Aim: Random sample imputation consists of taking random observation from the dataset and we use this observation to replace the nan values

When should it be used? 
It assumes that the data are missing completely at random(MCAR)

```python
import pandas as pd
df=pd.read_csv("D:/My Documents/Machine Learning/DataSets/titanic/train.csv", usecols=['Age','Fare','Survived','Embarked'])
df.head()
```

```python
def impute_nan(df,variable,mode):
    df[variable+"_mode"]=df[variable].fillna(mode)
```

```python
mode=df.Age.mode()
print(mode)
type(mode)
```

```python
mode = float(mode)
print(mode)
impute_nan(df,"Age",mode)
```

##### Advantages
1. Easy To implement
2. There is less distortion in variance

##### Disadvantage
1. Every situation randomness wont work

##### Capturing NAN values with a new feature

It works well if the data are not missing completely at random 

```python
df=pd.read_csv('titanic.csv', usecols=['Age','Fare','Survived'])
df.head()
```

```python
import numpy as np
df['Age_NAN']=np.where(df['Age'].isnull(),1,0)
```

```python
df.head()
```

```python
df.Age.median()
```

```python
df['Age'].fillna(df.Age.median(),inplace=True)
```
